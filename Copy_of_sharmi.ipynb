```python
# Install required libraries (only if not installed)
# !pip install scikit-learn pandas

# Import libraries
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Sample data (text and label)
texts = [
    "Government launches new education policy",
    "Aliens visited earth last night",
    "Scientists discover cure for common cold",
    "Fake news spreading about celebrity death",
    "NASA confirms water on moon",
    "Man claims he met a ghost in his backyard"
]

labels = [1, 0, 1, 0, 1, 0]  # 1 = Real, 0 = Fake

# Split data
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.3, random_state=42)

# Vectorize text data
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Train Logistic Regression model
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)

# Predict on test data
y_pred = model.predict(X_test_tfidf)

# Check accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Test with your own text
your_news = ["New vaccine developed for virus"]
your_news_tfidf = vectorizer.transform(your_news)
prediction = model.predict(your_news_tfidf)

print("Prediction (1=Real, 0=Fake):", prediction[0])
```

```python
# Install if needed
# !pip install scikit-learn pandas seaborn matplotlib

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
texts = [
    "Government launches new education policy",
    "Aliens visited earth last night",
    "Scientists discover cure for common cold",
    "Fake news spreading about celebrity death",
    "NASA confirms water on moon",
    "Man claims he met a ghost in his backyard"
]
labels = [1, 0, 1, 0, 1, 0]

# Split
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.3, random_state=42)

# TF-IDF
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Models to compare
models = {
    "Logistic Regression": LogisticRegression(),
    "Naive Bayes": MultinomialNB(),
    "SVM": LinearSVC()
}

accuracies = {}
conf_matrices = {}

# Train and evaluate
for name, model in models.items():
    model.fit(X_train_tfidf, y_train)
    y_pred = model.predict(X_test_tfidf)
    accuracies[name] = accuracy_score(y_test, y_pred)
    conf_matrices[name] = confusion_matrix(y_test, y_pred)

# Convert accuracy to DataFrame
acc_df = pd.DataFrame(list(accuracies.items()), columns=["Model", "Accuracy"])

#  Plot 1: Accuracy comparison bar chart
sns.barplot(data=acc_df, x="Model", y="Accuracy", palette="viridis")
plt.ylim(0, 1)
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy Score")
plt.show()

#  Plot 2: Confusion matrix heatmaps
for name, cm in conf_matrices.items():
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap="YlGnBu", xticklabels=['Fake', 'Real'], yticklabels=['Fake', 'Real'])
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```
